```python
from mgz import header, body
import os

with open('/Users/Parikshit_verma/Documents/GitHub/AgeofEmpires2_Voobly/rec.20191222-170345.mgz', 'rb') as data:
    eof = os.fstat(data.fileno()).st_size
    header.parse_stream(data)
    while data.tell() < eof:
        op = body.operation.parse_stream(data)
        if op.type == 'action' and op.action.type == 'build':
            print(op.action)
```
